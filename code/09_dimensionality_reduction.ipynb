## PCA (Principal Component Analysis)

```python
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
import numpy as np

# Load the faces dataset
faces, _ = fetch_olivetti_faces(return_X_y=True, shuffle=True, random_state=42)
n_samples, n_features = faces.shape
X = faces.copy()
print("Dataset size:", n_samples)
print("Number of features:", n_features)
```

```python
def show_images(images):
    pics_num = 6
    plt.figure(figsize=(10, 10))
    for i in range(pics_num):
        plt.subplot(1, 6, i+1)
        plt.imshow(images[i].reshape(64, 64), cmap='gray')
        plt.axis('off')
    plt.show()

show_images(X[:6])
```

```python
from sklearn.decomposition import PCA

# PCA with 0.8 variance
pca08 = PCA(n_components=0.8)
X_pca08 = pca08.fit_transform(X)
print("PCA features with 0.8 variance: ", X_pca08.shape[1])
show_images(pca08.inverse_transform(X_pca08[:6]))
```

```python
from sklearn.decomposition import PCA

# PCA with 0.95 variance
pca095 = PCA(n_components=0.95)
X_pca095 = pca095.fit_transform(X)
print("PCA features with 0.8 variance: ", X_pca095.shape[1])
show_images(pca095.inverse_transform(X_pca095[:6]))
```
